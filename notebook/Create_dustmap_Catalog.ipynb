```python
import healpy as hp
import numpy as np
import astropy.table as apTable
import tables_io
```

```python
sfd = hp.read_map('sfd_ebv.fits', nest=False)
lss_error = hp.read_map('lss_error.fits', nest=False)
```

```python
# Some lss error value are -1.6375e+30 indicating bad pixel. 
# These pixels of csfd use old values of sfd with no lss correction.
# Thus we are safe to use sfd uncertainty only for these pixels.
lss_error[lss_error < 0] = 0
```

```python
sfd_t = np.zeros(len(sfd), dtype=[("map", '>f8')])
sfd_t['map'] = sfd
sfd_table = apTable.Table(sfd_t)
sfd_table = tables_io.convert(sfd_table, tables_io.types.NUMPY_FITS)
tables_io.write(sfd_table, 'sfd_ebv', 'fit')
```

```python
lss_error_t = np.zeros(len(lss_error), dtype=[("map", '>f8')])
lss_error_t['map'] = lss_error
lss_error_table = apTable.Table(lss_error_t)
lss_error_table = tables_io.convert(lss_error_table, tables_io.types.NUMPY_FITS)
tables_io.write(lss_error_table, 'lss_error', 'fit')
```
