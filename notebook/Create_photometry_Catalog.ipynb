```python
from pathlib import Path
import opencosmo as oc
import numpy as np
import matplotlib.pyplot as plt
import pyarrow.parquet as pq
```

```python
# This tutorial assumes you are running on NERSC
all_catalogs_path = Path("/project/kicp/chto/roman-sompz/old/hlwas_cosmos_260120_UM_2026_01_22/")
available_catalogs = list(p.stem for p in all_catalogs_path.glob("*") if "latest" in p.stem)
print(available_catalogs)
```

```python
# Select one of the catalogs, and open the data
data_path = all_catalogs_path 
files = list(f for f in data_path.glob("*.hdf5") if f.stem.startswith("lc_cores"))
dataset = oc.open(*files)
```

```python
dataset.columns
```

```python
columns = ['lsst_u', 'lsst_g','lsst_r','lsst_i','lsst_z','lsst_y', 'roman_F106', 'roman_F129', 'roman_F158', 'redshift', 'ra', 'dec']
```

```python
data = dataset.select(columns).data
```

```python
data
```

```python
data = data.to_pandas()
data=data.rename(columns={"lsst_u": "flux_pgauss_LSST_u", 
                     "lsst_g": "flux_pgauss_LSST_g", 
                     "lsst_r": "flux_pgauss_LSST_r", 
                     "lsst_i": "flux_pgauss_LSST_i", 
                     "lsst_z": "flux_pgauss_LSST_z", 
                     "lsst_y": "flux_pgauss_LSST_y", 
                     'roman_F106':'flux_pgauss_Y',
                     'roman_F129':'flux_pgauss_J',
                     'roman_F158':'flux_pgauss_H', 'redshift': 'specz'}, errors="raise")
```

```python
m = (data['flux_pgauss_H']<24.5)&(data['ra']<2)&(data['dec']>84)&(data['dec']<89)
len(data['flux_pgauss_H'][m])
```

```python
data = data[m]
```

```python
for columns in data.columns:
    if 'flux' in columns:
        print(columns)
        data[columns] = 10**((22.5-data[columns])/2.5)
```

```python
for columns in data.columns:
    if 'flux' in columns:
        print(columns)
        carr=columns.split("_")
        carr[0]='flux_err'
        carr = "_".join(carr)
        data[carr]=1E-4
```

```python
data['ID']=np.arange(len(data))
```

```python
import pyarrow as pa
data= pa.Table.from_pandas(data)
```

```python
pq.write_table(data, '/project/kicp/chto/roman-sompz/hlwas_cosmos_260120_UM_2026_01_22.parquet')
```
